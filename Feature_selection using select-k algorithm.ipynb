```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
def select_k_algorithm(X,y,n):
    algorithm=SelectKBest(chi2,k=n)
    fit1=algorithm.fit(X,y)
    result=fit1.transform(X)
    return result
```

```python
def standarscalar(X,y):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,y_train,y_test
    
```

```python
def classification_test(classifier,X_test):
    y_pred=classifier.predict(X_test)
    Confution_Matrix=confusion_matrix(y_test,y_pred)
    Classification_Report=classification_report(y_test,y_pred)
    Accuracy=accuracy_score(y_test,y_pred)
    return classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix
```

```python
def LR(X_train,y_train,X_test):
    classifier=LogisticRegression(random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=classification_test(classifier,X_test)
    return classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix
```

```python
def KNN(X_train,y_train,X_test):
    classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=classification_test(classifier,X_test)
    return classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix
```

```python
def SVM(X_train,y_train,X_test):
    classifier=SVC(kernel="linear",random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=classification_test(classifier,X_test)
    return classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix
```

```python
def RF(X_train,y_train,X_test):
    classifier=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=classification_test(classifier,X_test)
    return classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix
```

```python
def DT(X_train,y_train,X_test):
    classifier=DecisionTreeClassifier(criterion="entropy",random_state=0)
    classifier.fit(X_train,y_train)
    classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=classification_test(classifier,X_test)
    return classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix
```

```python
X=df.drop("classification",axis=1)
y=df["classification"]
```

```python
select_k=select_k_algorithm(X,y,3)
log=[]
knn=[]
svm1=[]
rf=[]
dt=[]
```

```python
X_train,X_test,y_train,y_test=standarscalar(select_k,y)
```

```python
classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=LR(X_train,y_train,X_test)
log.append(Accuracy)
classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=KNN(X_train,y_train,X_test)
knn.append(Accuracy)
classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=SVM(X_train,y_train,X_test)
svm1.append(Accuracy)
classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=RF(X_train,y_train,X_test)
rf.append(Accuracy)
classifier,Accuracy,Classification_Report,X_test,y_test,Confution_Matrix=DT(X_train,y_train,X_test)
dt.append(Accuracy)
```

```python
dataframe=pd.DataFrame(columns=["LR","KNN","SVM","RF","DT"],index=["CHI_SQUARE"])
for i,j in enumerate(dataframe.index):
    dataframe["LR"][j]=log[0]
    dataframe["KNN"][j]=knn[0]
    dataframe["SVM"][j]=svm1[0]
    dataframe["RF"][j]=rf[0]
    dataframe["DT"][j]=dt[0]
```
